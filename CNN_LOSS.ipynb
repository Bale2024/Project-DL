```python


import pickle
from pickle import load
import os
import pandas as pd


path = r"C:\Users\USER\Desktop\pickle\metrics_c"
files = os.listdir(path)
cifar = []
for file in files:
  cifar.append(pickle.load(open(path + "\\" + file, "rb")))


```

```python
df_r = pd.DataFrame(cifar, index = ['Lenet5', 'Resnet18', 'Alexnet','VGGnet'])
df_r.transpose()
```

```python
df_flattened = pd.concat([pd.json_normalize(df_r[col]) for col in df_r], axis=1)
fin = df_flattened.transpose()
print(fin)

```

```python
import matplotlib.pyplot as plt
epochs = list(range(1,101))
plt.figure()
plt.title('Training Loss Curve')
plt.plot(epochs,fin[0], label='Lenet5')
plt.plot(epochs,fin[1], label='Resnet18')
plt.plot(epochs,fin[2], label='Alexnet')
plt.plot(epochs,fin[3], label='VGGnet')
#plt.xlabel('Epoch')
#plt.ylabel('Loss')
plt.legend()
plt.show()
```
