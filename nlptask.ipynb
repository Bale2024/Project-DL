```python
%%writefile app.py
from transformers import pipeline
import torch
import streamlit as st

# Functions defined to run the NLP tasks:

def predict_next_word(prompt, model, tokenizer, top_k=5):
    model.eval()
    # Tokenize input text
    inputs = tokenizer(prompt, return_tensors='pt')

    # Generate predictions
    with torch.no_grad():
        outputs = model(**inputs)

    # Get logits of the last token in the sequence
    next_token_logits = outputs.logits[:, -1, :]

    # Filter the top k tokens by their probability
    top_k_tokens = torch.topk(next_token_logits, top_k).indices[0].tolist()

    # Decode the top k tokens to their corresponding words
    predicted_tokens = [tokenizer.decode([token]) for token in top_k_tokens]

    return predicted_tokens

def generate_text(prompt, model, tokenizer, max_length=50):
    generated = tokenizer.encode(prompt, return_tensors='pt')

    model.eval()
    with torch.no_grad():
        for _ in range(max_length):
            outputs = model(generated)
            next_token_logits = outputs.logits[:, -1, :]
            next_token = torch.argmax(next_token_logits, dim=-1)
            generated = torch.cat((generated, next_token.unsqueeze(0)), dim=1)
            if next_token == tokenizer.eos_token_id:
                break

    return tokenizer.decode(generated[0], skip_special_tokens=True)

def chat_with_model(prompt, chat_history_ids, model, tokenizer):
  model.eval()
    # Encode the new user input, add the eos_token and return a tensor in PyTorch
  new_user_input_ids = tokenizer.encode(prompt + tokenizer.eos_token, return_tensors='pt')

    # Append the new user input tokens to the chat history
  bot_input_ids = torch.cat([chat_history_ids, new_user_input_ids], dim=-1) if chat_history_ids is not None else new_user_input_ids

    # Generate a response
  chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Decode the response and print it
  response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

  return response, chat_history_ids

def answer_question(question, context,model,tokenizer):
  # Tokenize the input question and context
  inputs = tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt")

  # Get input ids and attention mask
  input_ids = inputs["input_ids"].tolist()[0]

  # Get the model's output
  outputs = model(**inputs)
  answer_start_scores = outputs.start_logits
  answer_end_scores = outputs.end_logits

  # Get the most likely beginning and end of answer with the argmax of the score
  answer_start = torch.argmax(answer_start_scores)
  answer_end = torch.argmax(answer_end_scores) + 1

  # Convert tokens to the answer
  answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))

  return answer

def analyze_sentiment(text, model):
    result = model(text)
    return result[0]['label']

def summarize(text, model, tokenizer, max_length=130, min_length=30, do_sample=False):
  model.eval()
  inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
  summary_ids = model.generate(inputs, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4, early_stopping=True)
  summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
  return summary

def generate_image(prompt, pipe):
    with torch.no_grad():
        image = pipe(prompt).images[0]
    return image
#####  STREAMLIT DASHBOARD APP FOR NLP ##############
st.title("NLP Task Selector ")
task = st.selectbox("Select an NLP Task",["Text Summarization","Next Word Prediction","Story Completion","Chatbot","Sentiment Analysis","Question & Answering"])
input_text = st.text_area("Enter your text here")

if task == "Text Summarization":
  st.subheader("TEXT SUMMARIZATION")
  sum_bt = st.button("Summarize")
  if sum_bt:
    model = torch.load("summary.pth")
    tokenizer = torch.load("summary_tokenizer.pth")
    summary = summarize(input_text, model, tokenizer)
    st.write("Summary of given text:", summary)
elif task == "Next Word Prediction":
  st.subheader("NEXT WORD PREDICTION")
  nwp_bt = st.button("Predict")
  if nwp_bt:
    model = torch.load("GPT2.pth")
    tokenizer = torch.load("GPT2_tokenizer.pth")
    predicted_tokens = predict_next_word(input_text, model, tokenizer)
    st.write(f"The text entered is: {input_text}")
    st.write(f"Predicted words are:{'/'.join(predicted_tokens)}")
elif task == "Story Completion":
  st.subheader("Story Completion")
  generate_bt = st.button("Generate Story")
  if generate_bt:
    model = torch.load("GPT2.pth")
    tokenizer = torch.load("GPT2_tokenizer.pth")
    generated_story = generate_text(input_text, model, tokenizer)
    st.write(f"The text entered is: {input_text}")
    st.write(f"Generated story is: {generated_story}")
elif task == "Chatbot":
  st.subheader("CHATBOT")
  tokenizer = torch.load("chatbot_tokenizer.pth")
  model = torch.load("chatbot.pth")

  chat_history_ids = None
  if input_text:
    response, chat_history_ids = chat_with_model(input_text, chat_history_ids, model, tokenizer)
    st.write(f"The text entered is: {input_text}")
    st.write(f"Generated text is: {response}")
elif task == "Question & Answering":
  st.subheader("QUESTION & ANSWER")
  question = st.text_input("Enter a question")
  ans_bt = st.button("Answer")
  if ans_bt:
    model = torch.load("question_answering.pth")
    tokenizer = torch.load("question_answering_tokenizer.pth")
    answer = answer_question(question,input_text, model, tokenizer)
    st.write(f"The question entered is : {question}")
    st.write(f"Generated answer is: {answer}")
elif task == "Sentiment Analysis":
  st.subheader("SENTIMENT ANALYSIS")
  analyse = st.button("Analyse")
  if analyse:
    model = torch.load("sentiment_analysis.pth")
    result = model(input_text)
    st.write(f"The text entered is: {input_text}")
    st.write(f"Sentiment analysis result is: {result[0]['label']}")
elif task == "Image Generation":
  st.subheader("IMAGE GENERATION")
  img_bt = st.button("Generate")
  if img_bt:
    pipe = torch.load("image_generation.pth")
    image = generate_image(input_text, pipe)
    st.image(image, caption="Generated Image")
```
